```python
# expiry='22-Sep-2022'

dayE = 13
monthE ='Oct'
yearE = '2022'
```

```python
if dayE < 10 :
    val ='0'+str(dayE)
    expiry = val+"-"+monthE+"-"+yearE
else:
    expiry = str(dayE)+"-"+monthE+"-"+yearE
# expiry
```

```python
import requests
import pandas as pd

symbol =['BANKNIFTY','NIFTY']
oh_high=1000
oh_low=40

while True:
    try:
        for s in symbol:
            print(s,"OH")
            url='https://www.nseindia.com/api/quote-derivative?symbol={}'.format(s)
            headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:104.0) Gecko/20100101 Firefox/104.0'}

            r = requests.get(url,headers=headers)
            data=r.json()
            item = len(data['stocks'])
            list=[]
            timestamp=[]
            for i in range(item):
                list.append(data['stocks'][i]['metadata'])
                timestamp.append(data['opt_timestamp'])
            df_time=pd.DataFrame(timestamp)
            df_time.columns=['timestamp']

            df=pd.DataFrame(list)
            df["timestamp"]=df_time['timestamp']
            g1=df.groupby("expiryDate")
            df_expiry=g1.get_group(expiry)
            df_oh = df_expiry[(df_expiry['openPrice']==df_expiry['highPrice'])& (df_expiry['openPrice']>oh_low)& 
                              (df_expiry['openPrice']<oh_high)].sort_values(by="highPrice")
            df_oh_reorder = df_oh[['timestamp','optionType', 'strikePrice','openPrice', 'highPrice', 
                                   'lowPrice', 'closePrice','prevClose', 'lastPrice', 'change', 'pChange']]
            # df_oh_reorder.style.apply(blue_font, axis=0, subset=['openPrice','highPrice'])

            display(df_oh_reorder)
        break
    except:
        print("Network busy....trying again")
```
